```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import hvplot.pandas
import numpy as np
```

```python
# Reading in the 2021 and 2020 fantasy football data for running backs 
fantasy_football_2021 = pd.read_csv(
    "FantasyPros_Fantasy_Football_Statistics_RB.csv",
    index_col="Player"
)

fantasy_football_2020 = pd.read_csv(
    "FantasyPros_2020.csv",
    index_col="Player"
)
# Displaying the first 5 rows of the two dataframe 
display(fantasy_football_2021.head())
display(fantasy_football_2020.head())
```

```python
# Reorganizing the dataframes to display FPTS and FPTS/G
fantasy_points_2021 = fantasy_football_2021[["FPTS", "FPTS/G"]].dropna()
fantasy_points_2020 = fantasy_football_2020[["FPTS", "FPTS/G"]].dropna()

fantasy_points_combined = pd.concat([fantasy_points_2021, fantasy_points_2020])
fantasy_points_combined
```

```python
fantasy_data = fantasy_points_combined.groupby(["Player"]).mean()[["FPTS", "FPTS/G"]].sort_values("FPTS/G", ascending=False)

fantasy_data
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(fantasy_data)
```

```python
player_segments = model.predict(fantasy_data)

player_segments
```

```python
player_predictions = fantasy_data.copy()

player_predictions["Player Segment"] = player_segments

player_predictions.head()
```

```python
bad_running_backs = player_predictions[player_predictions['Player Segment'] == 0]

decent_running_backs = player_predictions[player_predictions['Player Segment'] == 2]

great_running_backs = player_predictions[player_predictions['Player Segment'] == 1]
```
