```python
import bs4 as bs
import urllib.request 
import pandas as pd
import numpy as np
```

```python
wk_01_source = urllib.request.urlopen("https://stathead.com/tiny/ckc9q").read()
soup = bs.BeautifulSoup(wk_01_source, 'lxml')
table = soup.find_all('table')
wk01_21_te = pd.read_html(str(table))[0]
wk01_21_te = wk01_21_te[['Team', 'Player', 'Week', 'Yds', 'TD', 'FantPt']].copy()
wk01_21_te.head()
```
