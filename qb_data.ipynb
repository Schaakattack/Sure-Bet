```python
import bs4 as bs
import urllib.request
import pandas as pd
```

```python
source = urllib.request.urlopen("https://www.pro-football-reference.com/years/2021/passing.htm").read()
soup = bs.BeautifulSoup(source, 'lxml')
```

```python
table = soup.find_all('table')
```

```python
df = pd.read_html(str(table))[0]
```

```python
df.head(10)
```

```python
df.to_csv("nfl.csv", index=False)
```
