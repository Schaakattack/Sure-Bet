```python
# Import necessary libraries
import bs4 as bs
import urllib.request 
import pandas as pd
import numpy as np
```

```python
# 2017 Tight End Data
source_2017 = urllib.request.urlopen("https://stathead.com/tiny/pw1Ap").read()
soup = bs.BeautifulSoup(source_2017, 'lxml')
table = soup.find_all('table')
te_2017 = pd.read_html(str(table))[0]
te_2017 = te_2017[['Date', 'Team', 'Player', 'Week', 'Yds', 'TD', 'FantPt']].copy()
te_2017.head()
```

```python
# 2018 Tight End Data
source_2018 = urllib.request.urlopen("https://stathead.com/tiny/47PSB").read()
soup = bs.BeautifulSoup(source_2018, 'lxml')
table = soup.find_all('table')
te_2018 = pd.read_html(str(table))[0]
te_2018 = te_2018[['Date', 'Team', 'Player', 'Week', 'Yds', 'TD', 'FantPt']].copy()
te_2018.head()
```

```python
# 2019 Tight End Data
source_2019 = urllib.request.urlopen("https://stathead.com/tiny/fTWUj").read()
soup = bs.BeautifulSoup(source_2019, 'lxml')
table = soup.find_all('table')
te_2019 = pd.read_html(str(table))[0]
te_2019 = te_2019[['Date', 'Team', 'Player', 'Week', 'Yds', 'TD', 'FantPt']].copy()
te_2019.head()
```

```python
# 2020 Tight End Data
source_2020 = urllib.request.urlopen("https://stathead.com/tiny/FiDAQ").read()
soup = bs.BeautifulSoup(source_2020, 'lxml')
table = soup.find_all('table')
te_2020 = pd.read_html(str(table))[0]
te_2020 = te_2020[['Date', 'Team', 'Player', 'Week', 'Yds', 'TD', 'FantPt']].copy()
te_2020.head()
```

```python
# 2021 Tight End Data
source_2021 = urllib.request.urlopen("https://stathead.com/tiny/PeWHC").read()
soup = bs.BeautifulSoup(source_2021, 'lxml')
table = soup.find_all('table')
te_2021 = pd.read_html(str(table))[0]
te_2021 = te_2021[['Date', 'Team', 'Player', 'Week', 'Yds', 'TD', 'FantPt']].copy()
te_2021.head()
```

```python
season_5yr = pd.concat(
    [te_2017, te_2018, te_2019, te_2020, te_2021], axis=0)

display(season_5yr.head(75))
display(season_5yr.tail(75))
```

```python
# Add collumn to dataframe to establish point threshold of postion
# for Tight End we want to have the equivalent of 2 touchdowns
season_5yr["Threshold"] = np.where(season_5yr["FantPt"] >= 12, 1, 0)

season_5yr.head(75)
```

```python
# Understand spread
season_5yr['Threshold'].value_counts()
```

```python
# Simplify date to year only
season_5yr.dtypes
```

```python
# Convert Date Column to year only
season_5yr['Date'] = pd.to_datetime(season_5yr['Date'])
season_5yr['Season'] = season_5yr['Date'].dt.year
season_5yr.dtypes
```

```python
# create unique id number for each player
season_5yr['ID'] = season_5yr.groupby(['Player']).ngroup() 
season_5yr
```

```python
# Create a copy of 5 Year Data
model_5yr = season_5yr.copy()
model_5yr
```

```python
# Drop Object Data and Re-order colums
model_5yr = model_5yr.drop(['Date', 'Team', 'Player'], axis=1)
model_5yr
```

```python
model_5yr = model_5yr[['ID', 'Threshold', 'Season', 'Week', 'Yds', 'TD', 'FantPt']]
model_5yr
```
